Load libraries

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

Load data

```python
breast_cancer_wisconsin_diagnostic = load_breast_cancer()
X = breast_cancer_wisconsin_diagnostic.data
y = breast_cancer_wisconsin_diagnostic.target
breast_cancer_wisconsin_diagnostic.target_names
```

Preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

Model seslection and training

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

Model testing

```python
y_pred = clf.predict(X_test)
#Prediction 1
print(y_pred[1])
print(y_test[1])
#Prediction 2
print(y_pred[30])
print(y_test[30])
#Prediction 3
print(y_pred[-1])
print(y_test[-1])
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=breast_cancer_wisconsin_diagnostic.target_names))
```

Visualization of the decision tree

```python
plt.figure(figsize=(20, 15))  # Increase figure size for better resolution
tree.plot_tree(clf, filled=True, feature_names=breast_cancer_wisconsin_diagnostic.feature_names, class_names=breast_cancer_wisconsin_diagnostic.target_names)
plt.title("Decision Tree for Breast Cancer  Prediction")
plt.savefig("decision_tree.png", dpi=300)  # Increase DPI for better resolution
plt.show()
```
